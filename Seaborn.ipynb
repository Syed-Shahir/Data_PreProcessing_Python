```python
#seaborn , and basic graphs 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#seaborn has some toy datasets for practice purpose
#use set them to have consistent them through the project
sns.set_theme()
```

```python
#lets import the tips dataset
tips = sns.load_dataset('tips')
# type(tips)
tips.head(5)
```

### Description: 
- A line plot displays data points connected by straight line segments. It is used to show trends over a period of time or another continuous variable.
### Use Case: 
- To visualize trends or changes over time

```python
#line plot
plt.figure(figsize=(6,4))
sns.lineplot(x = tips['total_bill'],y = tips['tip'],data = tips)

plt.title('Line plot')
plt.xlabel('Total Bill Payed')
plt.ylabel('Tip payed')
plt.show()
```

### Description:
- relplot (short for "relationship plot") is a high-level interface for creating relational plots in Seaborn. It combines the functionality of scatterplot and lineplot.
### Use Case:
- Use relplot when you want to create scatter or line plots with additional dimensions of data (e.g., faceting by rows and/or columns, adding hue for categorical variables, etc.)

```python
#relplot
plt.figure(figsize=(4,4))
#based on column it will produce multiple graphs based on different categories
#due to hue it will differentiate automatically with different colours based on the specified attribute
#now style will assign different representation based on specified attribute
#size parameter will differentiate between different categories of specified attribute with the help of different sizes
sns.relplot(x=tips['total_bill'],y=tips['tip'],col=tips['time'],hue=tips['smoker'],style=tips['smoker'],size=tips['size'])

plt.show()
```

```python
#import titanic dataset 
titanic = sns.load_dataset('titanic')
# type(titanic)
titanic.head(5)
```

### Description: 
- A count plot displays the count of observations in each categorical bin using bars. It is similar to a bar plot but specifically for counts of observations.

### Use Case: 
- To visualize the count of categories in a categorical variable.

```python
#count plot 
plt.figure(figsize = (4,4))
sns.countplot(x=titanic['class'],hue=titanic['sex'])

plt.ylim(0,400)

plt.show()
```

### Description: 
- A bar plot displays the distribution of a categorical variable. It shows the quantity of each category with rectangular bars.
### Use Case: 
- To compare the values of different categories.

```python
#bar plot
plt.figure(figsize=(4,4))
sns.barplot(x=titanic['class'],y=titanic['survived'])
plt.show()
```

### Description: 
- A histogram displays the distribution of a single continuous variable. It uses bins to divide the range of values into intervals and shows the frequency of values in each bin.
### Use Case: 
- To visualize the distribution of a variable.

```python
#histogram
plt.figure(figsize=(4,4))

sns.histplot(titanic['fare'],binwidth=10,bins=10)
plt.show()
```

```python
#distribution plot
plt.figure(figsize=(4,4))
sns.displot(titanic['fare'],binwidth=10,bins=10)
# sns.distplot(titanic['fare'])
plt.show()
```

### Description: 
- A box plot displays the distribution of a continuous variable through quartiles. It shows the median, upper and lower quartiles, and potential outliers.

### Use Case: 
- To visualize the distribution of a variable and detect outliers.

```python
# Box plot
plt.figure(figsize=(4,4))
sns.boxplot(x=iris["species"], y=iris["sepal_length"])
plt.show()
```

### Description: 
- A violin plot combines a box plot with a density plot. It provides a richer display of the distribution shape and density.
### Use Case: 
- To visualize the distribution of a variable and its density.

```python
#violin plot
plt.figure(figsize=(4,4))
sns.violinplot(x=iris["species"], y=iris["sepal_length"])
plt.show()
```

### Description: 
- A pair plot creates a matrix of scatter plots for all pairs of numerical variables in a DataFrame. It is useful for exploring relationships between multiple variables.
### Use Case: 
- To visualize pairwise relationships in a dataset.

```python
# Pair plot
sns.pairplot(iris, hue="species")
plt.show()
```

```python
#importing boston house dataset
#i got the dataset with me so i just have to load it here from local
file_path = 'HousingData.csv'
boston = pd.read_csv(file_path)
boston.head()
#now lets create a correlation so that we can visualize it with heatmap so that we can identify positive and negative 
correlation_matrix = boston.corr()
correlation_matrix
```

### Description:
- A heatmap displays data in a matrix format, with colors representing different values. It is useful for visualizing correlations between variables.
### Use Case:
- To show the correlation matrix or visualize data with color gradients.

```python
#now lets visulize it with heatmap
plt.figure(figsize=(10,4))
#square is to make it exact square
#annot is used to show the correlation coefficient value
#fmt is used to show the decimal places
#annot_kws is used for adjusting the annotation size
#cmap is used to determine the colour
sns.heatmap(correlation_matrix,annot=True,fmt = '.1f',annot_kws={'size':8},cmap = 'Greens')
plt.show()
```
