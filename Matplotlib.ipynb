## MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_axis = np.linspace(0,5,50)
sin_function = np.sin(np.pi*t_axis)
cos_function = np.cos(np.pi*t_axis)
```

```python
plt.figure(figsize=(6,4))
plt.plot(t_axis,sin_function,label='sin function')
plt.plot(t_axis,cos_function,label = 'cos function')
plt.xlabel("Time axis")
plt.ylabel("Output function")
plt.title("First graph")
plt.legend(loc = (0.7,0.8))
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.plot(t_axis,sin_function,'r+',label='sin function')
plt.plot(t_axis,cos_function,'g+',label = 'cos function')

plt.xlabel("Time axis")
plt.ylabel("Output function")
plt.title("Second Graph")

plt.xlim(-1,6)
plt.ylim(-2,2)

plt.grid(True)
plt.legend(loc = (0.7,0.8))
plt.show()
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))  # Adjusted figsize for better layout

# Plot in the first subplot
ax1.plot(t_axis, sin_function, 'r+', label='sin function')
ax1.set_xlabel("Time axis")
ax1.set_ylabel("Output function")
ax1.set_title("subplot 1")
ax1.set_xlim(-1, 6)
ax1.set_ylim(-2, 2)
ax1.grid(True)
ax1.legend(loc=(0.6, 0.9))

# Plot in the second subplot
ax2.plot(t_axis, cos_function, 'g+', label='cos function')
ax2.set_xlabel("Time axis")
ax2.set_ylabel("Output function")
ax2.set_title("subplot 2")
ax2.set_xlim(-1, 6)
ax2.set_ylim(-2, 2)
ax2.grid(True)
ax2.legend(loc=(0.6, 0.9))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#bar plot
y_values = [10,20,30,25,8,35]
x_values = ['sub1','sub2','sub3','sub4','sub5','sub6']

plt.figure(figsize=(6,4))
plt.bar(x_values,y_values,edgecolor = 'Black',color = 'Green')

# Add labels on top of bars
for i, v in enumerate(y_values): #enumerate returns index and value at that index
    #v is value but to adjust its position add or sub any value
    plt.text(i, v-5, str(v), ha='center', va='bottom', fontsize=12) #ha horizontal alignment , va verticle alignment

plt.ylim(0,40)
# plt.grid(True)

plt.xlabel('Subjects')
plt.ylabel('Values')
plt.title('Bar chart 2')
plt.tight_layout()
plt.show()
```

```python
#pie chart
plt.figure(figsize=(6,4))
plt.pie(y_values,labels=x_values)
plt.show()
```

```python
plt.figure(figsize=(6,4))
#autopct for the percentage
plt.pie(y_values,labels=x_values,autopct='%1.1f%%')
plt.show()
```

```python
import random as rd

# Generate random data for math and science scores
math_scores = np.random.randint(50, 100, size=30)
science_scores = math_scores + np.random.normal(0, 10, size=30)

# Display the first few data points
# print(math_scores[:5])
# print(science_scores[:5])

#Scatter plot
plt.figure(figsize=(6,4))
plt.scatter(math_scores,science_scores)

# Add titles and labels
plt.title("Scatter Plot", fontsize=16)
plt.xlabel("Math Scores", fontsize=14)
plt.ylabel("Science Scores", fontsize=14)

# Adding grid
# plt.grid(True)
plt.show()
```
