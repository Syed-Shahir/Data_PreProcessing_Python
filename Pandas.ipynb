```python
#now lets import data from our local 
file_path = 'HousingData.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
#lets say we want to check the information regarding the dataset
df.info()
```

```python
#if you want to know the dimensions of the dataset
df.shape
#the output means that we have got 506 rows and 14 columns
```

```python
#check if there are some null value or not 
df.isnull().sum()
#on the basis of output we can see that we have some null values 
```

```python
#count the number of values based on a certain column
df.value_counts('RAD')
```

```python
#lets say we want to group by the data based on some column
df1 = df.groupby('RAD').mean()
df1
#so we calculated the mean for each attribute based on grouping on a certain column
```

```python
#statistical data can be found manually aswell as all together
df.describe()
```

```python
#removing a particular column or row
#for row keep the axis = 0 and for column keep the axis = 1
df.drop(index=1,axis=0)#this will be temporary or you can save it in the other dataframe 
df_row = df.loc[0 : 2] # now first we will drop this than we will add it again 
# df_row.head()
df.drop(index = [0,1,2],axis=0,inplace=True) # to make it permenant we will write inplace = True
```

```python
df.head()
```

```python
#now lets add those rows back into our dataset
df = df._append(df_row,ignore_index=True)
```

```python
df.head()
```

```python
#now let suppose we want to remove the column 
df_col = df['B']
df.drop(columns='B',axis = 1,inplace=True)
df['B'] = df_col
df.head()
```
